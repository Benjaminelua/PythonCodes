# EDA in Pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\user\Downloads\world_population (2).csv")
df
```

```python
#Changing the decimals to 2 decimal place
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Find the number of null values per column
df.isnull().sum()
```

```python
#Checkinhg for how many unique values are in a column
df.nunique()
```

```python
#Looking at the top 10 largest countries by 2022 population
df.sort_values(by = '2022 Population', ascending = False).head(10)
```

```python
#Looking at correlations between the columns
corr_matrix = df[['2022 Population', '2020 Population', '2015 Population', '2010 Population', '2000 Population', '1990 Population', '1980 Population', '1970 Population', 'Area (km²)', 'Density (per km²)', 'Growth Rate', 'World Population Percentage']].corr()
corr_matrix
```

```python
#Visualising the correlations
sns.heatmap(corr_matrix, annot = True)
plt.rcParams['figure.figsize'] = (23,8)
plt.show()
```

```python
#Getting the highest popolation by average
df.groupby(['Continent'])[[ '1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population', ]].mean().sort_values(by= '2022 Population', ascending = False )
```

```python
df2 = df.groupby(['Continent'])[['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population']].mean().sort_values(by= '2022 Population', ascending = False )
df2
```

```python
#Transposing for better visualization
df3 = df2.transpose()
df3
```

```python
#Visualizing the population by average
df3.plot()
```
