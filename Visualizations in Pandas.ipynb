# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\user\Downloads\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('_classic_test_patch')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores', figsize = (8,4))
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
df.plot.barh( stacked = True, xlabel = 'Scores')
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 300, c = 'red')
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,4))
```

```python
df.plot.pie(y = 'Overall Rating', figsize = (13, 9))
```
